```python
NUMPY INDEXING AND SLICING OF  MATRIX
```

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
#for printing rows
arr[0]
```

```python
#for printing element in given matrix

arr[1,2]
```

```python
#for printing more the one element 
slice = arr[0:2,1:3]
slice
```

```python
slice = arr[3:,3:]
slice
```

```python
slice = arr[1:,2:]
slice
```

```python
# for column print 

slice = arr[:,2]
slice
```
